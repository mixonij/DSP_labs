```python
#Подключаем библиотеки
import pandas as pd
import numpy as np
```

```python
#Загружаем данные
data = pd.read_csv("data.csv", header = None)
data = data.iloc[:-1, range(0, 7)]
data
```

```python
# Усредненные значения
data_mean = pd.DataFrame(np.zeros((len(data), len(data.columns))))

for i in range(0, len(data.columns)):
    mean = np.mean(data[range(0,len(data.columns))[i]])
    
    for j in range(0, len(data)):
        data_mean.iloc[j, i] = data.iloc[j, i] - mean
        
data_mean
```

```python
# Корреляция
corr = pd.DataFrame(np.zeros((len(data.columns), len(data.columns))))

for i in range (0, len(data.columns)):
    for j in range(0, len(data.columns)):
        fg = np.sum(data_mean[i] * data_mean[j])
        f = np.sqrt(np.sum(data_mean[i] ** 2))
        g = np.sqrt(np.sum(data_mean[j] ** 2))
        corr.iloc[i, j] = fg / (f * g)
        
corr
```
