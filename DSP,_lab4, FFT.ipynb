```python
import wave
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq
```

```python
def show_graph(y):
    yf = fft(y)
    plt.figure(figsize=(10, 10))
    plt.plot(np.abs(yf)[1:60])
    plt.show()
```

```python
wr = wave.open('1.wav','r')

nchannels, sampwidth, framerate, nframes, comptype, compname =  wr.getparams()


binarydata = wr.readframes(nframes)

data = []
for i in range(nframes * 2):
    data.append(binarydata[i * 3 + 2])

left_1 = data[::2]

plt.plot(left_1[0:1000])
```
